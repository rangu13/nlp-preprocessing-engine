```python
## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love"
In NLP, "Love" and "love" are treated as different tokens due to case sensitivity. This can lead to duplication of meaning. Converting all text to lowercase ensures uniformity and avoids treating the same word as different tokens.

### 2. What happens if stopwords are not removed?
If stopwords (like "is", "the", "and") are not removed:
- The dataset becomes noisy
- Model performance may decrease
- Important words may get less weight

### 3. When removing stopwords can be harmful
- Sentiment Analysis: "I am not happy" → removing "not" changes meaning
- Question Answering: "What is your name?" → removing "what" loses intent

### 4. Stemming vs Lemmatization
- Stemming: Cuts words to root form (running → run, but sometimes incorrect)
- Lemmatization: Converts words to meaningful base form using vocabulary (running → run)
```

```python
import re
import nltk
from collections import Counter

nltk.download('punkt')
```

```python
def preprocess_text(text):
    try:
        if not text or text.strip() == "":
            return [], ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs and emails
        text = re.sub(r'http\S+|www\S+|https\S+|\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Handle repeated characters (soooo → so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove emojis and special characters
        text = re.sub(r'[^a-z\s]', '', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenize
        tokens = text.split()

        # Remove very short words except 'no', 'not'
        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except Exception as e:
        return [], ""
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 60)
```

```python
for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if tokens else 0

    print("Sentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", avg_length)
    print("-" * 60)
```

```python
## Token Analysis Observations

### 1. Sentence with most noise:
"Win $$$ now!!! Limited offer!!!"
Reason: Contains symbols, repeated punctuation, and promotional noise.

### 2. Sentence with most meaningful tokens:
"I am not happy with this"
Reason: After cleaning, meaningful sentiment words remain.
```

```python
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```
